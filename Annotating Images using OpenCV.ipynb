# Annotating Images

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
import matplotlib
matplotlib.rcParams['figure.figsize'] = (9.0, 9.0)
from PIL import Image
```

```python
## Reading Image in Colour
image = cv2.imread("C:/Users/BM/Desktop/MyProjects/03_Image_Annotation/Apollo_11_Launch.jpg", 1)

#Display the original image
plt.imshow(image[:, :, ::-1])
```

## Drawing a Line on an Image

```python
imageLine = image.copy()

# The line starts at (200, 100) and ends at (400, 100)
# The line color is Yellow and thickness is 5 px
# Line type is cv2.LINE_AA

cv2.line(imageLine, (200, 100), (400, 100), (0, 255, 255), thickness = 5, lineType = cv2.LINE_AA)
plt.imshow(imageLine[ :, :, ::-1])
```

## Drawing a Circle

```python
#Draw a Circle
imageCircle = image.copy()

# The line starts at (200, 100) and ends at (400, 100)
# The line color is Yellow and thickness is 5 px
# Line type is cv2.LINE_AA
cv2.circle(imageCircle, (900, 500), 100, (0, 0, 255), thickness = 5, lineType = cv2.LINE_AA)
plt.imshow(imageCircle[ :, :, ::-1])
```

## Drawing a Rectangle

```python
imageRect = image.copy()

# Top left corner of rectangle at (500, 50) and bottom right corner at (700, 600)
# The line color is Yellow and thickness is 5 px
# Line type is cv2.LINE_AA

cv2.rectangle(imageRect, (500, 50), (700, 600), (255, 0, 255), thickness = 5, lineType = cv2.LINE_AA)
plt.imshow(imageRect[ :, :, ::-1])
```

## Adding text to Images

```python
imageText = image.copy()
text = "Apollo 11 Saturn V lauch, July 16, 1969"
fontScale = 2.3
fontFace = cv2.FONT_HERSHEY_PLAIN
fontColor = (0, 255, 0)
fontThickness = 2

cv2.putText(imageText, text, (200, 700), fontFace, fontScale, fontColor, fontThickness, cv2.LINE_AA )

# Display the image
plt.imshow(imageText[:, :, ::-1])
```
