## Splitting and Merging Color Channels

```python
# Split the image into B,G,R components.
img_NZ_bgr = cv2.imread("C:/Users/BM/Desktop/MyProjects/New_Zealand_Lake.jpg", 1)

# Image in BGR format according to OpenCV
plt.imshow(img_NZ_bgr)

b,g,r = cv2.split(img_NZ_bgr)

#Show the channels
plt.figure(figsize = [20,5])
plt.subplot(141); plt.imshow(r, cmap = 'gray'); plt.title("Red Channel")
plt.subplot(142); plt.imshow(g, cmap = 'gray'); plt.title("Green Channel")
plt.subplot(143); plt.imshow(b, cmap = 'gray'); plt.title("Blue Channel")

#Merge the individual channels into a BGR image
imgMerged = cv2.merge((b, g, r))
# Show the merged output
plt.subplot(144); plt.imshow(imgMerged[:, :, : :-1]); plt.title("Merged Image")
```

```python
# Alternatively Merge the individual channels into a RGB image
img_RGB = cv2.merge((r,g,b))

# Display the RGB image
plt.imshow(img_RGB)
```

# Converting to different Colour Spaces

## Changing from BGR to RGB

```python
img_NZ_rgb = cv2.cvtColor(img_NZ_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_NZ_rgb)
```

## Chane to HSV ColorSpace

```python
img_NZ_hsv = cv2.cvtColor(img_NZ_bgr, cv2.COLOR_BGR2HSV)

# Split the HSV Channels
h,s,v = cv2.split(img_NZ_bgr)

#Show the channels
plt.figure(figsize = [20,5])
plt.subplot(141); plt.imshow(h, cmap = 'gray'); plt.title("Hue Channel")
plt.subplot(142); plt.imshow(s, cmap = 'gray'); plt.title("Saturation Channel")
plt.subplot(143); plt.imshow(v, cmap = 'gray'); plt.title("Value Channel")
plt.subplot(144); plt.imshow(img_NZ_rgb, cmap = 'gray'); plt.title("Original Image")
```

## Modifying Individual Channel

```python
h_new = h+10
img_NZ_merged = cv2.merge((h_new,s,v))
img_NZ_RGB = cv2.cvtColor(img_NZ_merged, cv2.COLOR_HSV2RGB)

#Show the channels
plt.figure(figsize = [20,5])
plt.subplot(141); plt.imshow(h_new, cmap = 'gray'); plt.title("Hue Channel")
plt.subplot(142); plt.imshow(s, cmap = 'gray'); plt.title("Saturation Channel")
plt.subplot(143); plt.imshow(v, cmap = 'gray'); plt.title("Value Channel")
plt.subplot(144); plt.imshow(img_NZ_RGB, cmap = 'gray'); plt.title("Modified Image")
```
